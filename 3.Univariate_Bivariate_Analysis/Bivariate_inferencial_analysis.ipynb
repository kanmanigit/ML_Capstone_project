```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/Users/kanmaniramasamy/Documents/HopeAI_Tamil/week10_ML_DatascienceAssignment_Capstone_project/heart.csv")
dataset
```

```python
# Separation of Quan and Qual columns
quan_columns = []
qual_columns = []
for columnName in dataset.columns:
    if dataset[columnName].dtype=="object":
        qual_columns.append(columnName)
    else:
        quan_columns.append(columnName)
    
 
```

```python
#Plot histogram for each numeric features

for col in dataset[quan_columns]:
   
    plt.figure() # create a new fig for each plot
    plt.hist(dataset[col])
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
### #Plot pdf for each numeric features

for col in dataset[quan_columns]:
    sns.distplot(dataset[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Find Skewness of all the numerical feature..

for col in dataset[quan_columns]:
    sk=dataset[col].skew()
    print(f'Skew of {col}: ',sk)
    
```

## Bivariant analysis

```python
### Bivariant Analysis
# does Cholestrol  affect the Heart Disease ? (Numerical -numerical)
sns.scatterplot(x=dataset["Cholesterol"],y=dataset["HeartDisease"])
```
