```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
dataset = pd.read_csv("heart.csv")
dataset
```

```python
#finding the quatitative columns for covariant calculation

quan=[]
qual=[]
for columnName in dataset.columns:
    if dataset[columnName].dtypes=='O':
        qual.append(columnName)
    else:
        quan.append(columnName)
```

```python
quan
```

## Covariant on the numerical column
Covariance is a way to measure how two variables change together. If two variables have a positive covariance, they tend to increase or decrease together. A negative covariance means that when one variable increases, the other tends to decrease. 

```python
dataset[quan].cov()
```

## Observarion:
examples
1. Age and RestingBP (44.43): Positive, indicating that as age increases, resting blood pressure also tends to increase slightly.
2. Age and Cholesterol (-98.31): Negative, so as age increases, cholesterol tends to decrease, but the large negative value suggests this relationship may vary widely.
3. Age and FastingBS (0.79): Positive but close to zero, indicating almost no relationship between age and fasting blood sugar.
4. RestingBP and Cholesterol (204.32): Positive, suggesting that higher resting blood pressure is associated with higher cholesterol.
5. RestingBP and HeartDisease (0.99): Positive but low, showing that people with higher resting blood pressure might have a slightly higher chance of heart disease.



## Correlation on the numerical column
showing how strongly each pair of variables is related on a scale from -1 to 1. 
1:  Perfect positive correlation; when one variable increases, the other also increases proportionally.
-1: Perfect negative correlation; when one variable increases, the other decreases proportionally.
0: No correlation; the variables do not have any linear relationship.

```python
dataset[quan].corr()
```

## Observarion:
example
1. Age and RestingBP (0.254): Small positive correlation; as age increases, resting blood pressure tends to increase slightly.
2. Age and MaxHR (-0.382): Moderate negative correlation; as age increases, maximum heart rate tends to decrease.
3. Cholesterol and MaxHR (0.236): Small positive correlation; higher cholesterol is weakly associated with higher max heart rate.
4. Cholesterol and HeartDisease (-0.233): Small negative correlation, meaning higher cholesterol is associated with slightly lower heart disease risk here, which might be counterintuitive and could depend on specific factors in this dataset.
